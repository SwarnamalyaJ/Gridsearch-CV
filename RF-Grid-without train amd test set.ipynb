```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python

dataset.columns
```

```python
indep=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
indep
```

```python
depn=dataset[["charges"]]
depn
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV 
param_grid={'criterion':["squared error","friedman_mse","absolute error","poisson"],'max_features':['None','sqrt','log2'],'n_estimators':[10,50,100]}
grid=GridSearchCV(RandomForestRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
```

```python
grid.fit(indep,depn)
```

```python
GridSearchCV(estimator=RandomForestRegressor(), n_jobs=-1,
             param_grid={'criterion': ["squared error","friedman_mse","absolute error","poisson"],
                         'max_features': ['None', 'sqrt', 'log2'],
                         'n_estimators': [10, 50,100]},
             verbose=3)
```

```python
re=grid.cv_results_
print=("best_parameter{}:".format(grid.best_params_))
```
