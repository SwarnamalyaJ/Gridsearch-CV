```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male','smoker_yes']]
independent
```

```python
dependent=dataset[["charges"]]
dependent
```

```python
from sklearn.model_selection import GridSearchCV 
from sklearn.tree import DecisionTreeRegressor
param_grid={'criterion':["squared error","friedman_mse","absolute error","poisson"],'max_features':[None,'sqrt','log2'],'splitter':['best','random']}
```

```python
grid=GridSearchCV(DecisionTreeRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(independent,dependent)
```

```python
best_model = grid.best_estimator_
```

```python
y_pred = best_model.predict(independent)
```

```python
re=grid.cv_results_
print=("best_parametes{}:".format(grid.best_params_))
```

```python

print
```
